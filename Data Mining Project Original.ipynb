# **-----------FORM HERE, WE WILL MAKE DUMMY VARIABLES AND MODELS-------------**

```
data['success'] = np.where(data['Succeeded']>0, 1, 0)
data
```

```
dfdummy=pd.get_dummies(data.Route,prefix='Route')
dfdummy
```

```
data1=pd.concat([data,dfdummy],axis=1)
data1
```

```
y=data1['success']
x=data1.drop(columns=['success','Succeeded','Success Percentage','Route','Date'])
```

```
x
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

***----------------------Using LOGISTIC REGRESSION------------------------***

```
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(X_train,y_train)

logistic.score(X_train,y_train)
```

```
logistic = LogisticRegression()
logistic.fit(X_train,y_train)
logistic.score(X_train,y_train)
```

***----------------------Using SVM------------------------***

```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, y_train)
```

```
y_pred = svclassifier.predict(X_test)
```

```
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```
svclassifier.score(X_train,y_train)
```
